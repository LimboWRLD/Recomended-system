```python
import pandas as pd
from surprise import Dataset, Reader, SVD
from surprise.model_selection import train_test_split
from surprise.accuracy import rmse

ratings = pd.read_csv('../datasets/ml-latest-small/ratings.csv')
movies = pd.read_csv('../datasets/ml-latest-small/movies.csv')

print("Ratings Data:")
print(ratings.head())

print("\nMovies Data:")
print(movies.head())
```

```python
import matplotlib.pyplot as plt

ratings['rating'].value_counts().sort_index().plot(kind='bar', title='Ratings Distribution')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()

movies['genres'] = movies['genres'].str.split('|')
genres = movies.explode('genres')
genres_count = genres['genres'].value_counts()

genres_count.plot(kind='bar', title='Number of Movies per Genre')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.show()
```
